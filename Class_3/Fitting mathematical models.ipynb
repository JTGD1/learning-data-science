```python
max_complexity = 1
for complexity in range(1, max_complexity+1):
    # Get complex x
    X = data_tools.X(complexity)
    
    # Model
    model = LogisticRegression()
    model.fit(X, Y)
    
    # Plot
    plt.subplot(1, max_complexity, complexity)
    data_tools.Decision_Surface(X, Y, model)
plt.tight_layout()
plt.show()
```
